```python
"""
Practica 3 | Tries de codificación

Angeles Rojo Sofia
Magno Gomez Ricardo Leon

"""

# Bibliotecas

import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

# Variables

excel = input("Agrega la ubicacion del excel con el alfabeto deseado:")
tabla = pd.read_excel(excel, dtype = str)
cod = dict(zip(tabla['Letras'], tabla['Cadenas']))
dec = dict(zip(tabla['Cadenas'], tabla['Letras']))

# Clases

class NodoTrie:
  """
  Clase NodoTrie indica si es el final de la palabra o referencia a
  un nodo hijo. Es la estructura basica para cada nodo dentro del
  Trie.
  """
  def __init__(self):
    """
    Metodo constructor de la clase NodoTrie.
    """
    self.omega = False
    self.alpha = {}

class Trie:
  """
  Clase que simula una estructura Trie o Arbol.
  """
  def __init__(self, excel):
    """
    Metodo constructor de la clase Trie.
    Recibe un excel con el que se trabaja.
    """
    self.raiz = NodoTrie()
    self.excel = excel

  def hojas(self, palabra):
    """
    Recibe una palabra y la inserta en el Trie. Recorre cada letra
    empleando nodos hijos de ser necesario.
    """
    nodo = self.raiz
    for i in palabra:
      if i not in nodo.alpha:
        nodo.alpha[i] = NodoTrie()
      nodo = nodo.alpha[i]
    nodo.omega = True

  def codificacion(self, cadena):
    """
    Recibe una cadena de texto que, con el uso del alfabeto dado,
    la codificara y regresara un codigo binario.
    """
    no = [a for a in cadena if a not in cod]
    if no:
      error = ("ERROR. Los siguientes caracteres no están en el "
                          "alfabeto: " + ", ".join(no))
      print(error)
      return ""
    return ' '.join(cod.get(a, "") for a in cadena)


  def decodificacion(self, codigo_binario):
    """
    Recibe un codigo binario y, conforme al alfabeto dado, regresa
    la decodificacion del codigo en texto.
    """
    codigo_binario = codigo_binario.replace(" ","")
    lista = ""
    lista_palabras = []
    for i in codigo_binario:
      lista += i
      if lista in dec:
        lista_palabras.append(lista)
        lista = ""
      else:
        pass

    if lista != "":
      print("REVISA EL CODIGO DADO, HAY ALGO QUE ANDA MAL.")
      print(f"EXISTEN DATOS ERRONEOS EN EL CODIGO, LO QUE SE LOGRO DECODIFICAR"
            f" FUE:")
      print("")

    lista_palabras_terminadas = " ".join(lista_palabras)
    caracteres = lista_palabras_terminadas.split(" ")
    palabra = ""

    for b in caracteres:
      caracter = dec.get(b, "")
      palabra += caracter

    return palabra

# Funciones

def practica_3():
  """
  Funcion encargada de ejecutar toda la practica 3.
  """
  print(tabla)
  print("------------------------------------------------------------------")
  print("                         ¡Bienvenido!")
  print("Para continuar, introduzca un mensaje que quiera codificar.")
  print("Todo lo que ingreses debe poderse escribir con el alfabeto.")
  print("Si por contrario, quiere decodificar un codigo, de enter.")
  print("")
  mensaje = input("Introduzca el mensaje que quiera coficar: ")
  print("")

  mi_trie = Trie(tabla)

  if mensaje != "":
    numero_binario = mi_trie.codificacion(mensaje)
    if numero_binario:  # Se verifica que numero_binario no esté vacío
      binario_unido = numero_binario.replace(" ", "")
      print("La palabra", mensaje, "en binario es:", binario_unido)

  if mensaje == "":
    codigo_binario = input(f"Introduzca su codigo binario que quiera"
                           f" decodificar: ")
    print("")
    texto = mi_trie.decodificacion(codigo_binario)
    print("El codigo binario proporcionado se ve como:",texto)

  print("------------------------------------------------------------------")

# Main

if __name__ == "__main__":
  practica_3()
```
